```python
#Step-1
import pandas as pd
import numpy as np

rno = 102303998
filepath= 'data.csv'

ar = 0.05 * (rno % 7)

br = 0.3 * ((rno % 5) + 1)

print(f"Roll Number: {rno}")
print(f"Calculated ar: {ar}")
print(f"Calculated br: {br}")

df = pd.read_csv(filepath,encoding='latin1')

feature_col='no2'

df_clean = df.dropna(subset=[feature_col]).copy()
x = df_clean[feature_col].values

z = x + ar * np.sin(br * x)
df_clean['z'] = z
print("\nTransformation complete.")
print(df_clean[[feature_col, 'z']].head())
```

```python
#Step-2
import numpy as np

mu_hat = np.mean(z)

sigma_sq_hat = np.var(z)

lambda_hat = 1 / (2 * sigma_sq_hat)
c_hat = 1 / (np.sqrt(sigma_sq_hat) * np.sqrt(2 * np.pi))

print("--- Learned Parameters ---")
print(f"Mean (μ)      : {mu_hat:}")
print(f"Lambda (λ)    : {lambda_hat:}")
print(f"Constant (c)  : {c_hat:}")
```
